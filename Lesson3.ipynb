### Вычисление длины вектора ###

```python
import math
def length(x, y, z):
    return math.sqrt(x**2 + y**2 + z**2)
    
print(f"Длина вектора [2, 7, 3] равна {round(length(2, 7, 3), 2)}")
```

### Построение графика окружности ###

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(-2, 2, 100000)
y = np.sqrt(4-x**2)
y2 = -np.sqrt(4-x**2)
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
```

### Построение графика эллипса ###

```python
x = np.linspace(-2, 2, 100000)
y = np.sqrt(1 - (x**2)/4)
y2 = -np.sqrt(1 - (x**2)/4)
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
```

### Плоскости в пространстве ### 

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_title('Параллельные плоскости') 
X = np.arange(-5, 5, 1)
Y = np.arange(-5, 5, 1)
X, Y = np.meshgrid(X, Y)
Z2 = -(10*X+2*Y)/2
Z1 = -(5*X+Y+10)
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_title('Эллипсоид') 
X = np.arange(-10, 10, 0.01)
Y = np.arange(-10, 10, 0.01)
X, Y = np.meshgrid(X, Y)
Z = 6*np.sqrt(1-(X**2)/90-(Y**2)/40)
Z2 = -6*np.sqrt(1-(X**2)/90-(Y**2)/40)
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)

# Как избавиться от ошибки Runtimewarning???
```

### Перевод полярных координат в декартовы

```python
def decart(r, a):
    x = r*cos(a)
    y = r*sin(a)
    return f'({x, y})'
r = int(input('Введите полярный радиус '))
a = int(input('Введите полярный угол '))
print(f"Координаты в декартовой системе {decart(r, a)}")
```

### Системы уравнений и неравенств 

```python
x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.plot(x, x**2 - 1)
```

```python
from scipy.optimize import fsolve
def f(p):
    x, y = p
    return (np.exp(x) + x*(1 - y) - 1, x**2 - 1 - y)

x1, y1 = fsolve(f, (2, 2))
print(x1, y1)
```
